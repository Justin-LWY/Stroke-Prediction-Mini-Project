```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy import stats
sb.set()
```

```python
data = pd.read_csv('train.csv')
```

## 1a)

```python
sale = pd.DataFrame(data['SalePrice']) #predictor
grliv = pd.DataFrame(data['GrLivArea']) #result
```

```python
g_train, g_test, s_train, s_test = train_test_split(grliv, sale, test_size=0.2)

linreg = LinearRegression()
linreg.fit(g_train, s_train)
```

```python
print('The coefficients: a =', linreg.coef_)
```

```python
s_training_pred = linreg.predict(g_train)

f, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].title.set_text('Training set')
axes[0].scatter(g_train, s_train)
axes[0].plot(g_train, s_training_pred, color='r')
    
axes[1].title.set_text('Testing set')
axes[1].scatter(g_test, s_test)
axes[1].plot(g_test, s_testing_pred, color='r')
plt.show()
```

## 1b)

```python
s_testing_pred = linreg.predict(g_test)

mse = mean_squared_error(s_test, s_testing_pred)
print("Goodness of fit metric:")
print("Explained Variance (R^2):", linreg.score(g_test, s_test), "\n")

print("Prediction accuracy metric:")
print("Root Mean Squared Error (RMSE):", np.sqrt(mse))
```

I use RMSE as opposed to MSE as the latter is reliant on the number of total data points that are available. RMSE allows for the representation in terms of relevant values. In this case, RMSE is the error (deviation in prediction) in **SalePrice** represented in dollars.

## 2)

```python
def univariate(pred):
    x_train, x_test, s_train, s_test = train_test_split(pred, sale, test_size=0.2)

    linreg = LinearRegression()
    linreg.fit(x_train, s_train)

    s_training_pred = linreg.predict(x_train)
    s_testing_pred = linreg.predict(x_test)

    f, axes = plt.subplots(1, 2, figsize=(16,8))
    axes[0].title.set_text('Training set')
    axes[0].scatter(x_train, s_train)
    axes[0].plot(x_train, s_training_pred, color='r')
    
    axes[1].title.set_text('Testing set')
    axes[1].scatter(x_test, s_test)
    axes[1].plot(x_test, s_testing_pred, color='r')
    plt.show()

    train_meanse = mean_squared_error(s_train, s_training_pred)
    test_meanse = mean_squared_error(s_test, s_testing_pred)
    
    print("------------Training set------------")
    print("Goodness of fit metric:")
    print("Explained Variance (R^2):", linreg.score(x_train, s_train), "\n")

    print("Prediction accuracy metric:")
    print("Root Mean Squared Error (RMSE):", np.sqrt(train_meanse), "\n")
    
    print("------------Testing set------------")
    print("Goodness of fit metric:")
    print("Explained Variance (R^2):", linreg.score(x_test, s_test), "\n")

    print("Prediction accuracy metric:")
    print("Root Mean Squared Error (RMSE):", np.sqrt(test_meanse))
```

```python
bsmt = pd.DataFrame(data['TotalBsmtSF'])
garage = pd.DataFrame(data['GarageArea'])
```

```python
univariate(bsmt)
```

```python
univariate(garage)
```

Using the *testing sets* of all 3 variables:

Firstly, between all 3 *Goodness of Fit*, the closest value to 1 is **GrLivArea** at around 0.5 explained variance metric.

Secondly, looking at *Prediction Accuracy*, using the Root Mean Squared Error metric, **GrLivArea** still has a smaller margin of error compared to the other 2.

Based on the respective metrics, **GrLivArea** seems to be the most accurate when predicting **SalePrice** among the three.

## 3a)

```python
housedata = pd.DataFrame(data[['GrLivArea', 'SalePrice']])

for i in housedata:
    q1 = housedata[i].describe(include='all').loc['25%']
    q3 = housedata[i].describe(include='all').loc['75%']
    IQR = q3 - q1
    btmThresh = q1 - 1.5 * IQR
    topThresh = q3 + 1.5 * IQR
    outliers = housedata[(housedata[i] < btmThresh) | (housedata[i] > topThresh)]
    cleanset = housedata.drop(outliers.index)
    


f = plt.figure(figsize=(16,8))
plt.title("Plot after cleaning dataset")
plt.scatter(data=cleanset, x='GrLivArea', y='SalePrice')
plt.show()
```

## 3b)

```python
cleangr = cleanset.iloc[:, :1]
cleansale = cleanset.iloc[:, 1:]
```

```python
cg_train, cg_test, cs_train, cs_test = train_test_split(cleangr, cleansale, test_size=0.2)

cleanlinreg = LinearRegression()
cleanlinreg.fit(cg_train, cs_train)
```

```python
print('The coefficients: a =', cleanlinreg.coef_)
```

```python
cs_training_pred = cleanlinreg.predict(cg_train)
cs_testing_pred = cleanlinreg.predict(cg_test)

f, axes = plt.subplots(1, 2, figsize=(16,8))
axes[0].title.set_text('Training set')
axes[0].scatter(cg_train, cs_train)
axes[0].plot(cg_train, cs_training_pred, color='r')
    
axes[1].title.set_text('Testing set')
axes[1].scatter(cg_test, cs_test)
axes[1].plot(cg_test, cs_testing_pred, color='r')
plt.show()
```

## 3c)

```python
ctrain_meanse = mean_squared_error(cs_train, cs_training_pred)
ctest_meanse = mean_squared_error(cs_test, cs_testing_pred)
    
print("------------Training set------------")
print("Goodness of fit metric:")
print("Explained Variance (R^2):", linreg.score(cg_train, cs_train), "\n")

print("Prediction accuracy metric:")
print("Root Mean Squared Error (RMSE):", np.sqrt(ctrain_meanse), "\n")
    
print("------------Testing set------------")
print("Goodness of fit metric:")
print("Explained Variance (R^2):", linreg.score(cg_test, cs_test), "\n")

print("Prediction accuracy metric:")
print("Root Mean Squared Error (RMSE):", np.sqrt(ctest_meanse))
```

The **clean data set** would be a **better model** than the **raw model** in *Problem 1*.

By comparing both the *testing set* metrics,

The *Goodness of Fit*, as seen from the explained variance metric, has gone down with the clean data set. The *Prediction Accuracy*, looking at RMSE, has also gone down pretty significantly.

I decided that the clean data set would be a better model as I weight the *Prediction Accuracy* heavier than *Goodness of Fit*. The main objective of a model is to accurately predict **SalePrice** which the clean data set does better when looking at RMSE.
