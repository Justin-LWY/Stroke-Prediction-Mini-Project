```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import scipy.stats as stats
sb.set()
```

```python
trainData = pd.read_csv('train.csv')
houseData = pd.DataFrame(trainData[['LotArea', 'GrLivArea', 'TotalBsmtSF', 'GarageArea', 'SalePrice']])
```

## 1a)

```python
f, axes = plt.subplots(2, 2, figsize=(20, 15))

sb.boxplot(data=houseData["LotArea"], orient="h", color="r", ax=axes[0, 0])
sb.boxplot(data=houseData["GrLivArea"], orient="h", color="b", ax=axes[1, 0])
sb.boxplot(data=houseData["TotalBsmtSF"], orient="h", color="g", ax=axes[0, 1])
sb.boxplot(data=houseData["GarageArea"], orient="h", color="y", ax=axes[1, 1])
```

```python
for i in houseData:
    q1 = houseData[i].describe(include='all').loc['25%']
    q3 = houseData[i].describe(include='all').loc['75%']
    IQR = q3 - q1
    btmThresh = q1 - 1.5 * IQR
    topThresh = q3 + 1.5 * IQR
    outliers = houseData[(houseData[i] < btmThresh) | (houseData[i] > topThresh)]

    print('Number of outliers for', i, ':', len(outliers))
```

Therefore we can conclude that **LotArea** has the most outliers at **69**

## 1b)

```python
houseData.describe()
```

```python
for i in houseData:
    mean = houseData[i].describe(include='all').loc['mean']
    median = houseData[i].describe(include='all').loc['50%']
    skew = mean - median
    print(i,'Skew:',skew)
```

```python
for i in houseData:
    res = stats.skewtest(houseData[i])
    print(i, 'Skew:', res.statistic)
```

**LotArea** has the largest **positive** skew among the variables

## 1c)

```python
f = plt.figure(figsize=(12,12))
sb.heatmap(houseData.corr(), annot=True)

houseData.corr()
```

We can tell here that the two variables that have the highest correlation with SalePrice is **GrLivArea** and **GarageArea**.

```python
sb.jointplot(data=houseData, x="SalePrice", y="GrLivArea", hue="SalePrice")
sb.jointplot(data=houseData, x="SalePrice", y="GarageArea", hue="SalePrice")
```

The two variables that will help the most in predicting SalePrice is **GrLivArea** and **GarageArea**

## 2a)

```python
housedf = pd.DataFrame(trainData[["MSSubClass", "Neighborhood", "BldgType", "OverallQual", "SalePrice"]])
```

```python
sb.catplot(data=housedf, y="MSSubClass", kind="count")
sb.catplot(data=housedf, y="Neighborhood", kind="count")
sb.catplot(data=housedf, y="BldgType", kind="count")
sb.catplot(data=housedf, y="OverallQual", kind="count")
```

We can tell that the variable with the **most levels** is **Neighborhood**

The **level with the most houses** is under the variable **BldgType:** **1Fam**.

## 2b)

```python
f, axes = plt.subplots(2, 2, figsize=(20,15))

sb.boxplot(data=housedf, x="MSSubClass", y="SalePrice", color="r", ax=axes[0, 0])
sb.boxplot(data=housedf, x="Neighborhood", y="SalePrice", color="b", ax=axes[1, 0])
sb.boxplot(data=housedf, x="BldgType", y="SalePrice", color="g", ax=axes[0, 1])
sb.boxplot(data=housedf, x="OverallQual", y="SalePrice", color="y", ax=axes[1, 1])
```

The variables Neighborhood and MSSubClass do not seem to have a discernable pattern with SalePrice.

SalePrice seems to increase as **OverallQual** increases.

SalePrice does not seem to vary greatly between **BldgType** variables.

```python
f = plt.figure( figsize=(15,25))

sb.boxplot(data=housedf, y="SalePrice", x="BldgType", hue="OverallQual")
```

Observe that the SalePrice of the house is related to the **OverallQual**. As OverallQual increases, the SalePrice generally increases. Comparing alongside **BldgType** is helpful as the variance in SalePrice of the variable is low.
